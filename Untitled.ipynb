Суть задачи состояла в том чтобы предугадать уровень гравитации в конкретном участке поля, в качестве датасета был дан field_train.csv, где у нас есть координаты x и y а также гравитация в этой точке, в качестве решение получилось довольно лаконичным и показало метрику 14.77 / 15 на закрытом тестовом датасете.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('field_train.csv')
train
```

```python
example = pd.read_csv('field_example.csv')
```

```python

plt.figure(figsize=(10, 8))
scatter = plt.scatter(train['longitude'], train['latitude'], c=train['intensity'], cmap='viridis', s=50, alpha=0.6)


if 'intensity' in train.columns:
    plt.colorbar(scatter, label='Intensity Level')

plt.title('Распределение точек по координатам')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python

scaler = StandardScaler()
X_scaled = scaler.fit_transform(train[['longitude', 'latitude']])
y = train['intensity'].values

knn = KNeighborsRegressor(n_neighbors=5, weights='distance')  
knn.fit(X_scaled, y)


grid_x, grid_y = np.meshgrid(np.linspace(train['longitude'].min(), train['longitude'].max(), 100),
                             np.linspace(train['latitude'].min(), train['latitude'].max(), 100))
X_new = np.column_stack((grid_x.flatten(), grid_y.flatten()))
X_new_scaled = scaler.transform(X_new)


y_pred = knn.predict(X_new_scaled)

synthetic_df = pd.DataFrame({
    'longitude': X_new[:, 0],
    'latitude': X_new[:, 1],
    'intensity': y_pred
})


mask = (synthetic_df['longitude'] >= train['longitude'].min()) & (synthetic_df['longitude'] <= train['longitude'].max()) & \
       (synthetic_df['latitude'] >= train['latitude'].min()) & (synthetic_df['latitude'] <= train['latitude'].max())
synthetic_df = synthetic_df[mask]

print(f"Количество синтезированных точек: {len(synthetic_df)}")

full_df = pd.concat([train, synthetic_df], ignore_index=True)
full_df = full_df.drop_duplicates(subset=['longitude', 'latitude'])  

print(f"Общее количество точек: {len(full_df)}")


plt.figure(figsize=(12, 8))
scatter = plt.scatter(full_df['longitude'], full_df['latitude'], c=full_df['intensity'], 
                      cmap='viridis', s=30, alpha=0.6)
plt.colorbar(scatter, label='Intensity Level')
plt.title('Полная карта точек (исходные + синтезированные с k-NN)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python

example_scaled = scaler.transform(example[['longitude', 'latitude']])

y_pred_example = knn.predict(example_scaled)

example_with_intensity = example.copy()
example_with_intensity['intensity'] = y_pred_example
```
